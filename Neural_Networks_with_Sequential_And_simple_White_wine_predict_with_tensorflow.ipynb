```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
print(x_train.shape)
print(y_train.shape)
```

```python
x_train=x_train.reshape(-1,28*28).astype("float32")/255.0
x_test=x_test.reshape(-1,28*28).astype("float32")/255.0
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#Sequential API
```

```python
model=keras.Sequential(
    [
        keras.Input(shape=(28*28)),
        layers.Dense(512,activation='relu'),
        layers.Dense(256,activation='relu'),
        layers.Dense(10),
    ]
)
print(model.summary())
model.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=keras.optimizers.Adam(lr=0.001),
    metrics=["accuracy"]
)
model.fit(x_train,y_train,batch_size=32,epochs=5,verbose=2)
model.evaluate(x_test,y_test,batch_size=32,verbose=2)
```

```python
import pandas as pd
df=pd.read_csv("white_wine.csv")
```

```python
def clean_dataset(dataset):
  dataset["Year"] = dataset.Year.replace("N.V.","2016").astype(int)
  return dataset
clean_data = clean_dataset(df)
clean_data
```

```python
df = clean_data.copy()
df["Year"] = df.Year.replace("N.V.","2016").astype(int)
df.Name = df["Name"].astype('category').cat.codes
df.Country = df["Country"].astype('category').cat.codes
df.Region = df["Region"].astype('category').cat.codes
df.Winery = df["Winery"].astype('category').cat.codes
```

```python
# from sklearn.model_selection import train_test_split
# dr = ["Price"]
# x=df.drop(dr, axis=1)
# y = df["Price"]
# X_train, X_validation, Y_train, Y_validation = train_test_split(x, y, test_size=0.20, random_state=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop('Price', axis=1), df['Price'], test_size=0.2, random_state=42)
```

```python
import numpy as np
inputs=keras.Input(shape=(784,))
x=layers.Dense(512,activation='relu')(inputs)
x=layers.Dense(256,activation='relu')(x)
outputs=layers.Dense(10,activation='softmax')(x)
model2=keras.Model(inputs=inputs,outputs=outputs)
model2=keras.Model(inputs=inputs,outputs=outputs)
model2.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    optimizer=keras.optimizers.Adam(lr=0.001),
    metrics=["accuracy"]
)



model2.fit(x_train,y_train)

```

```python
pred=model2.predict(x_test)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score

# Assuming pred contains your model predictions
# If your predictions are probabilities, you might want to convert them to class labels
# For example, if pred is the output of model.predict(X_test), you can use np.argmax
y_pred_labels = np.argmax(pred, axis=1)

# Assuming y_test contains your true labels
# If your labels are in one-hot encoded format, you can use np.argmax to get class labels
y_test_labels = y_test

# Calculate accuracy
accuracy = accuracy_score(y_test_labels, y_pred_labels)

print(f"Accuracy: {accuracy * 100:.2f}%")

```
